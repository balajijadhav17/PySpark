# withColumn Method

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("withColumn Methos").getOrCreate()
```

```python
from pyspark.sql.functions import *
from datetime import date
data = [(1,"rahul",date(2026, 10, 1),"18000"),
         (2,"payal",date(2026, 10, 5),"20000"), 
         (3,"komal",date(2026, 11, 13),"25000"), 
         (4,"rani",date(2026, 11, 20),"30000"),
         (5,"swati",date(2026, 12, 30),"40000")]
schema = ["id","name","admission_date","Fees"]
df=spark.createDataFrame(data=data,schema=schema)
df.show()
df.printSchema()
```

### change data type of the column

```python
df = df.withColumn("Fees",col("Fees").cast("Integer"))
df.show()
df.printSchema()
```

```python
df = df.withColumn("id",col("id").cast("Integer"))
df.show()
df.printSchema()
```

### create new constant column

```python
df = df.withColumn("Country",lit("India"))
df.show()
```

```python
display(df)
```

### update existing column data

```python
df= df.withColumn("Fees",col("Fees")+5000)
df.show()
```

```python
df.printSchema()
```

### Copy colume

```python
df=df.withColumn("copy_salary_column",col("Fees"))
df.show()
```
