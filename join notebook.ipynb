# join

```python
data1 = [(1,"balaji","male","20",2),
         (2,"ramkishan","male","21",1), 
         (3,"lavkush","male","23",4)]
schema1 = ["emp_id","emp_name","gender","age","dept_id"]

data2 = [(1,"IT"),(2,"HR"),(3,"FINANCE")]
schema2 = ["dept_id","dept_name"]
empDF = spark.createDataFrame(data1, schema1)
deptDF = spark.createDataFrame(data2, schema2)
empDF.show()
deptDF.show()

```

## inner join

```python
empDF.join(deptDF,empDF.dept_id == deptDF.dept_id,"inner").show()
```

## left join

```python
empDF.join(deptDF,empDF.dept_id == deptDF.dept_id,"left").show()
```

## right join

```python
empDF.join(deptDF,empDF.dept_id == deptDF.dept_id,"right").show()
```

## full join

```python
empDF.join(deptDF,empDF.dept_id == deptDF.dept_id,"full").show()
```

## leftsemi

```python
empDF.join(deptDF,empDF.dept_id == deptDF.dept_id,"leftsemi").show()
```

## leftanti

```python
empDF.join(deptDF,empDF.dept_id == deptDF.dept_id,"leftanti").show()
```

## self join

```python
data = [(1,"balaji","male","20",2),
         (2,"ramkishan","male","21",1), 
         (3,"lavkush","male","23",4)]
schema = ["emp_id","emp_name","gender","age","manager_id"]

df=spark.createDataFrame(data,schema)
df.show()
```

```python
from pyspark.sql.functions import col
df.alias("empData").join(df.alias("managerData"),\
                  col("empData.manager_id") == col("managerData.emp_id"),\
                  "left")\
                .select(col("empData.emp_name").alias("empName"),col("managerData.emp_name").alias("managerName")).show()
```
