# Timestamp

```python
from pyspark.sql.functions import current_timestamp,to_timestamp,lit,hour,minute,second
df = spark.range(2)
df.show()
df1 = df.withColumn("currentTimestamp", current_timestamp())
df1.show(truncate=False)
df1.printSchema()

df2 = df1.withColumn("timestampInstring",lit('12.25.2022 08:12:24'))
df2.show()
df2.printSchema()


df3 = df2.withColumn("stringToTimestamp",to_timestamp(df2.timestampInstring,'MM.dd.yyyy HH:mm:ss'))
df3.show(truncate=False)
df3.printSchema()


df4 = df1.select("*",hour(df1.currentTimestamp).alias("hour"),\
    minute(df1.currentTimestamp).alias("minute"),\
    second(df1.currentTimestamp).alias("second"))
df4.show(truncate=False)
df4.printSchema()
```
