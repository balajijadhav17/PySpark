```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("Write DataFrame Into Json File").getOrCreate()
```

```python
row =  [(1,"balaji"),(2,"suresh"),(3,"ramesh")]
columns = ["id", "name"]
df= spark.createDataFrame(data = row,schema = columns)
display(df)
df.printSchema()
```

```python
df.write.json("/Volumes/pyspark/pyspark_schema/volume2/friend/")
```

```python
%sql
create volume pyspark.pyspark_schema.volume2;
```

```python
d=spark.read.json("/Volumes/pyspark/pyspark_schema/volume2/friend/")
display(d)
d.printSchema()
```

```python
df1=spark.read.format("json")\
    .option("header",True)\
    .option("inferSchema",True)\
    .load("/Volumes/pyspark/pyspark_schema/volume2/friend/")
df1.show()
display(df1)
```
