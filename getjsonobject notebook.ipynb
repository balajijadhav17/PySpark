# get_json_object

```python
data = [
    (1,'maheer', '{"address":{"city":"hyd","state":"telengana"},"gender":"male"}'),
    (2,'wafa',   '{"address":{"city":"banglore","state":"karnataka"},"gender":"male"}')
]

schema = ["id",'name', 'props']

df = spark.createDataFrame(data, schema)
df.show(truncate=False)
df.printSchema()
```

```python
from pyspark.sql.functions import get_json_object
df.select("id","name",get_json_object("props","$.gender").alias("gender")).show(truncate=False)
```

```python
from pyspark.sql.functions import get_json_object
df.select("id","name",get_json_object("props","$.address.city").alias("city"),get_json_object("props","$.address.state").alias("state")).show(truncate=False)
```
