# row_number,rank,dense_rank

```python
from pyspark.sql.functions import col
data = [(1,"balaji","jadhav","male","18000","IT"),
         (2,"ramkishan","jadhav","male","20000","HR"), 
         (3,"lavkush","pawar","male","25000","Sales"), 
         (4,"alok","pawar","male","30000","Marketing"),
         (5,"gopal","rathod","male","40000","IT"),
         (6,"mishti","more","female","40000","HR"),
         (7,"aiti","aade","female","50000","Sales")]
schema = ["id","firstname","lastname","gender","salary","department"]
df=spark.createDataFrame(data=data,schema=schema)
df = df.withColumn("salary",col("salary").cast("integer"))
df.show()
df.printSchema()
```

```python
from pyspark.sql.functions import row_number,rank,dense_rank
from pyspark.sql.window import Window

df.sort('department').show()
```

```python
window = Window.partitionBy('department').orderBy('salary')
df.withColumn('rownumber',row_number().over(window)).show()
```

```python
w = Window.orderBy('salary')
df.withColumn('rownumber',row_number().over(w)).show()
```

```python
w = Window.orderBy('salary')
df.withColumn('rank',rank().over(w)).show()
```

```python
w = Window.orderBy('salary')
df.withColumn('denserank',dense_rank().over(w)).show()
```

```python
window = Window.partitionBy('department').orderBy('salary')
df.withColumn('rank',rank().over(window)).show()
df.withColumn('denserank',dense_rank().over(window)).show()
```
