# DATAFRAME

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName(
    'DataFrame')\
    .getOrCreate()
```

```python
from pyspark.sql import Row
df = spark.createDataFrame([
	Row(id=1, name="balaji"),
	Row(id=2, name="ramkishan"),
	Row(id=3, name="lavkush"),
    Row(id=4, name="alok"),
    Row(id=5, name="gopal")
])
df.show()
```

# dataframe using python list method1

```python
l1=[(1,"balaji"),(2,"ramkishan"),(3,"lavkush"),(4,"alok"),(5,"gopal")]

df=spark.createDataFrame(data=l1,schema=["id","name"])
df.show()
df.printSchema()
```

```python
display(df)
```

# dataframe using python list method2

```python
data=[{"id":1,"name":"balaji"},
    {"id":2,"name":"ramkishan"},
    {"id":3,"name":"lavkush"},
    {"id":4,"name":"alok"},
    {"id":5,"name":"gopal"}]

df=spark.createDataFrame(data)
df.show()
```

# creating a dataframe by zipping the two lists

```python
data1 = [1,2,3,4,5]
data2 = ["balaji","ramkishan","lavkush","alok","gopal"]
column = ["id","name"]
data=zip(data1,data2)
df=spark.createDataFrame(data = data , schema= column)
df.show()
```

# dataframe using python dictionary

```python
data=[{"id":1,"name":"balaji"},
    {"id":2,"name":"ramkishan"},
    {"id":3,"name":"lavkush"},
    {"id":4,"name":"alok"},
    {"id":5,"name":"gopal"}]
df=spark.createDataFrame(data)
df.show()
```

```python
l2={"id":[1,2,3,4,5],"name":["balaji","ramkishan","lavkush","alok","gopal"]}

data = list(zip(l2["id"], l2["name"]))

df1=spark.createDataFrame(data,schema=["id","name"])
df1.show()
```

# By Definining Schema defination

```python
from pyspark.sql.types import *

data = [(1,"balaji"),(2,"ramkishan"),(3,"lavkush"),(4,"alok"),(5,"gopal")]

schema = StructType([
    StructField("id", IntegerType(), True),
    StructField("name", StringType(), True)])

df = spark.createDataFrame(data=data,schema=schema)
df.show()
df.printSchema()
```
