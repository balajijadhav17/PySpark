# Date

```python
df = spark.range(2)
df.show()
```

```python
from pyspark.sql.functions import current_date
df1 = df.withColumn("CurrentDate", current_date())
df1.show()
df1.printSchema()
# datetype default format yyyy-mm-dd
```

```python
from pyspark.sql.functions import date_format
df2=df1.withColumn("CurrentDate", date_format(df1.CurrentDate,"yyyy/mm/dd"))
df3=df1.withColumn("CurrentDate", date_format(df1.CurrentDate,"yyyy.mm.dd"))
df2.show()
df2.printSchema()
df3.show()
df3.printSchema()
```

```python
from pyspark.sql.functions import to_date
df4 = df2.withColumn("currentDate",to_date(df2.CurrentDate,"yyyy/mm/dd"))
df4.show()
df4.printSchema()
```
