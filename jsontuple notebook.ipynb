# json_tuple

```python
data = [("1",'{"firstname":"balaji", "lastname":"jadhav"}'),("2",'{"firstname":"lavkush", "lastname":"pawar"}')]
schema = ["id","name"]
df = spark.createDataFrame(data, schema)
df.show(truncate=False)
df.printSchema()
```

```python
from pyspark.sql.functions import json_tuple
df.select("name",json_tuple(df.name,"firstname","lastname").alias("firstname","lastname")).show(truncate=False)
```
