# select 

```python
data1 = [(1,"balaji","jadhav","male","20","IT"),
         (2,"ramkishan","jadhav","male","21","HR"), 
         (3,"lavkush","pawar","male","23","Sales"), 
         (4,"alok","pawar","male","22","Marketing"),
         (5,"gopal","rathod","male","40000","IT"),
         (6,"mishti","more","female","40000","HR"),
         (7,"aiti","aade","female","40000","Sales")]

schema1 = ["id","firstname","lastname","gender","age","department"]
df1=spark.createDataFrame(data1,schema1)

display(df1)
```

### select single column

```python
df1.select("firstname").show()
df1.select(df1.firstname).show()
df1.select(df1["firstname"]).show()
```

### select multiple columns

```python
df1.select("firstname","lastname").show()
df1.select(df1.firstname,df1.lastname).show()
df1.select(df1["firstname"],df1["lastname"]).show()
```

```python
df1.select(["id","department"]).show()
```

### select all column of the dataframe

```python
df1.select("*").show()
```

### select with col

```python
from pyspark.sql.functions import col
df1.select(col("firstname"),col("lastname")).show()
```

```python
df1.select(col("*")).show()
```
