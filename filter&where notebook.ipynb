## filter function 

```python
from datetime import date
data = [(1,"balaji","jadhav","male",date(2026, 10, 1),"18000"),
         (2,"ramkishan","jadhav","male",date(2026, 10, 5),"20000"), 
         (3,"lavkush","pawar","male",date(2026, 11, 13),"25000"), 
         (4,"alok","pawar","male",date(2026, 11, 20),"30000"),
         (5,"gopal","rathod","male",date(2026, 12, 30),"40000"),
         (6,"mishti","more","female",date(2026, 12, 2),"40000"),
         (7,"aiti","aade","female",date(2026, 12, 28),"40000")]
schema = ["id","firstname","lastname","gender","admission_date","Fees"]
df=spark.createDataFrame(data=data,schema=schema)
df.show(truncate=False)
df.printSchema()
```

```python
df.filter(df.gender == "male").show()
```

```python
df.filter("gender = 'female'").show()
```

```python
display(df.filter(df.lastname == "jadhav") and df.filter(df.Fees < 20000))
```

## where function

```python
df.where(df.gender == "male").show()
df.where("gender = 'female'").show()
```

```python
df.where((df.gender.like("%a%")) & (df.Fees == 40000)).show()
```
