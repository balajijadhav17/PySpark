# to_json

```python
data = [("1",{"firstname":"balaji", "lastname":"jadhav"}),("2",{"firstname":"lavkush", "lastname":"pawar"})]
schema = ["id","name"]
df = spark.createDataFrame(data, schema)
df.show(truncate=False)
df.printSchema()
```

```python
from pyspark.sql.functions import to_json
df1=df.withColumn("info",to_json(df.name))
df1.show(truncate=False)
df1.printSchema()
```
