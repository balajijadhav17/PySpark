# partitionBy

```python
data =[("1","balaji","male","IT"),("2","alok","male","HR"),("3","gopal","male","sales"),
       ("4","anushka","female","IT"),("5","payal","female","HR"),("6","sapna","female","sales")]
schema = ["id","name","gender","department"]
df = spark.createDataFrame(data = data, schema = schema)
df.show()
```

## partitionBy department

```python
df.write \
  .mode("overwrite") \
  .partitionBy("department") \
  .parquet("/Volumes/pyspark/pyspark_schema/volume1/outputtemp1")
# df.write.parquet("/Volumes/pyspark/pyspark_schema/volume1/outputtemp2/",mode="overwrite",partitionBy="gender")
```

## access partition data 

```python
df1 = spark.read.parquet("/Volumes/pyspark/pyspark_schema/volume1/outputtemp1/department=IT")
df1.show()
df1.printSchema()
```

## partitionBy gender

```python
df.write \
    .mode("overwrite") \
    .partitionBy("gender") \
    .parquet("/Volumes/pyspark/pyspark_schema/volume1/outputtemp2")
```

```python
df2=spark.read.parquet("/Volumes/pyspark/pyspark_schema/volume1/outputtemp2/gender=female")
df2.show()
df2.printSchema()
```

# partitionBy with multiple columns

```python
df.write \
  .mode("overwrite") \
  .partitionBy(["department","gender"]) \
  .parquet("/Volumes/pyspark/pyspark_schema/volume1/outputtemp3")
```

```python
df5 = spark.read.parquet("/Volumes/pyspark/pyspark_schema/volume1/outputtemp3")
df5.show()
df5.printSchema()
```

```python
df4=spark.read.parquet("/Volumes/pyspark/pyspark_schema/volume1/outputtemp3/department=IT/gender=female")
df4.show()
df4.printSchema()
```
