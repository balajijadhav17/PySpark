# Date

```python
from pyspark.sql.functions import datediff,months_between,add_months,date_add,year,month
from pyspark.sql import functions as F
df = spark.createDataFrame(
    [('2026-06-11','2026-11-11')],
    ["d1","d2"]
)

df.show()
df.printSchema()

df.withColumn("dateDifference", datediff(df.d1, df.d2)).show()
df.withColumn("monthDifference", months_between(df.d1, df.d2)).show()
df.withColumn("addMonths", add_months(df.d2,3)).show()
df.withColumn("subMonths", add_months(df.d2,-3)).show()
df.withColumn("addDate", date_add(df.d2,3)).show()
df.withColumn("subDate", date_add(df.d2,-3)).show()
df.withColumn("year", year(df.d2)).show()
df.withColumn("month", month(df.d2)).show()
df.withColumn("day_name", F.date_format("d2", "EEEE")).show()
df.withColumn("month_name", F.date_format("d2", "MMMM")).show()
```
