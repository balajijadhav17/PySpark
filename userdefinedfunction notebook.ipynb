```python
data = [(1,"balaji",1000,100),(2,"ramkishan",2000,200)]
schema = ["id","name","salary","bonus"]
df = spark.createDataFrame(data,schema)
df.show()
```

```python
def totalpay(s,b):
    return s+b
from pyspark.sql.functions import udf
from pyspark.sql.types import IntegerType
totalpayment = udf(lambda s,b: totalpay(s,b), IntegerType())
# df.select(df.name,totalpayment(df.salary,df.bonus).alias("totalpay")).show())
```

```python
df.withColumn("totalpay",totalpayment(df.salary,df.bonus)).show()
```

```python
from pyspark.sql.types import IntegerType
@udf(returnType=IntegerType())
def totalpay(s,b):
    return s+b
```

```python
df.select("*",totalpay(df.salary,df.bonus).alias("totpay")).show()
```

```python
data = [(1,"balaji",1000,100),(2,"ramkishan",2000,200)]
schema = ["id","name","salary","bonus"]
df = spark.createDataFrame(data,schema)
df.createOrReplaceTempView("emps")
```

```python
%sql
select * from emps;
```

```python
def totalpay(s,b):
    return s+b
spark.udf.register(name="totalpaysql",f=totalpay,returnType=IntegerType() )
```

```python
%sql
select id,name ,salary,bonus, totalpaysql(salary,bonus) as totalPays  from emps
```
