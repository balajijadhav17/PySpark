```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Write dataframe into parquet").getOrCreate()
```

```python
row =  [(1,"balaji"),(2,"suresh"),(3,"ramesh")]
columns = ["id", "name"]
df= spark.createDataFrame(data = row,schema = columns)
display(df)
df.printSchema()
```

```python
df.write.parquet("/Volumes/pyspark/pyspark_schema/volume2/friend1/")
```
