# groupBy

```python
from datetime import date
data = [(1,"balaji","jadhav","male","18000","IT"),
         (2,"ramkishan","jadhav","male","20000","HR"), 
         (3,"lavkush","pawar","male","25000","Sales"), 
         (4,"alok","pawar","male","30000","Marketing"),
         (5,"gopal","rathod","male","40000","IT"),
         (6,"mishti","more","female","40000","HR"),
         (7,"aiti","aade","female","40000","Sales")]
schema = ["id","firstname","lastname","gender","salary","department"]
df=spark.createDataFrame(data=data,schema=schema)
display(df)
```

```python
from pyspark.sql.functions import col
df=df.withColumn("salary",col("salary").cast("int"))
df.printSchema()
```

```python
df.groupBy("department").count().show()
print()
df.groupBy("department").max("salary").show()
print()
df.groupBy("department").avg("salary").show()
print()
df.groupBy("department").min("salary").show()
print()
df.groupBy("department").sum("salary").show()
```
