```python
from pyspark.sql import SparkSession
spark=SparkSession.builder.appName("Column Operation").getOrCreate()
```

```python
from datetime import date
data = [(1,"balaji","jadhav",date(2026, 10, 1),"18000"),
         (2,"ramkishan","jadhav",date(2026, 10, 5),"20000"), 
         (3,"lavkush","pawar",date(2026, 11, 13),"25000"), 
         (4,"alok","pawar",date(2026, 11, 20),"30000"),
         (5,"gopal","rathod",date(2026, 12, 30),"40000")]
schema = ["id","firstname","lastname","admission_date","Fees"]
df=spark.createDataFrame(data=data,schema=schema)
df.show(truncate=False)
df.printSchema()
```

### Access  columns

```python
df.select(df["Fees"]).show()
```

```python
df.select(df["id"],df["Fees"]).show()
```

```python
from pyspark.sql.functions import col
df.select(col("id")).show()
```

```python
df.select(col("id"),col("Fees")).show()
```

```python
df.select("firstname","lastname").show()
```

### add constant column

```python
from pyspark.sql.functions import lit
df=df.withColumn("new_col",lit("1"))
display(df)
```

```python
data = [("balaji","M","kolhapur"),("ramkishan","M","amravati"),("lavkush"," ","pune"),("pari","F","nashik"),
        ("pinki","F","nanded")]
schema=["name","gender","city"]
df1=spark.createDataFrame(data,schema)
df1.show()
df1.printSchema()
```

# alias

```python
df1=df1.select(df1.gender,df1.city,df1.name.alias("student_name"))
display(df1)
```

```python
df1=df1.select(df1.student_name,df1.gender.alias("student_sex"),df1.city.alias("student_city"))
df1.show()
```

# Ascending And Descending

```python
# ascending
df1.sort(df1.student_name.asc()).show()
```

```python
# descending
df1.sort(df1.student_name.desc()).show()
```

### lit

```python
df1=df1.withColumn("college_code",lit(563904))
display(df1)
df1.printSchema()
```

# cast

```python
df1 = df1.select(df1.student_name,df1.student_sex,df1.student_city,df1.college_code.cast("long"))
df1.show()
df1.printSchema()
```

# Like

```python
df1.filter(df1.student_name.like("p%")).show()
```
