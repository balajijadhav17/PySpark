# Fill

```python
from datetime import date
data = [(1,"balaji","jadhav",None,18000,None),
         (2,"ramkishan","jadhav","male",20000,"HR"), 
         (3,"lavkush","pawar","male",25000,"Sales"), 
         (4,"alok","pawar","male",None,"Marketing"),
         (5,"gopal","rathod","male",40000,"IT"),
         (6,"mishti","more",None,40000,"HR"),
         (7,"aiti",None,"female",40000,"Sales")]
schema = ["id","firstname","lastname","gender","salary","department"]
df=spark.createDataFrame(data=data,schema=schema)
df.show()
df.printSchema()
```

### find total number of null values in each columns

```python
from pyspark.sql.functions import count, when,col

df.select([
    count(when(col(c).isNull(), c)).alias(c)
    for c in df.columns
]).show()
```

```python
df.na.fill("Unknow").show() # fill all null with Unknow by default all columns
```

```python
df.na.fill("Unknow",["gender"]).show() 
```

```python
df.na.fill("Unknow",["gender","department","lastname"]).show() 
```

# fillna

```python
df.fillna("Unknow").show()  # fill all null with Unknow by default all columns
```

```python
df.fillna("Unknow",["gender","department","lastname"]).show() 
```

```python
df.fillna(0, subset=["salary"]).show()
```

```python
df.fillna({"salary": 0}).show()
```

## fill with mean

```python
from pyspark.sql.functions import mean

mean_value = df.select(mean("salary")).collect()[0][0]
print(mean_value)
df.fillna({"salary": mean_value}).show()
```

## fill with median 

```python
median_value = df.approxQuantile("salary", [0.5], 0.0)[0]
print(median_value)
df.fillna({"salary": median_value}).show()
```

## fill with mode

```python
from pyspark.sql.functions import col

mode_value = (
    df.groupBy("lastname")
      .count()
      .orderBy(col("count").desc())
      .first()[0]
)
print(mode_value)

df.fillna({"lastname": mode_value}).show()
```
