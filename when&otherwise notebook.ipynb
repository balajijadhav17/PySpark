# When and Otherwise

```python
data = [("balaji","male","kolh"),("ramkishan","male","amrav"),("lavkush","male","pu"),("pari","female","nash"),
        ("pinki","female"," ")]
schema=["name","gender","city"]
df=spark.createDataFrame(data,schema)
df.show()
df.printSchema()
```

```python
from pyspark.sql.functions import col,when
df=df.withColumn("identity",when(col("gender") == "male","boy").otherwise("girl"))
df.show()
df.printSchema()
```

```python
from pyspark.sql.functions import when
df = df.select(
    col("name"),
    col("gender"),
    col("identity"),
    when(col("city") == "kolh", "Kolhapur")
    .when(col("city") == "amrav", "Amravati")
    .when(col("city") == "pu", "Pune")
    .when(col("city") == "nash", "Nashik")
    .otherwise("Unknown")
    .alias("city")
)
df.show()
df.printSchema()
```
