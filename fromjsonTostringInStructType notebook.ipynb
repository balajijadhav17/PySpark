# from_json() function to convert json string into StructType

```python
data = [("1",'{"firstname":"balaji", "lastname":"jadhav"}'),("2",'{"firstname":"lavkush", "lastname":"pawar"}')]
schema = ["id","name"]
df = spark.createDataFrame(data, schema)
df.show(truncate=False)
df.printSchema()
```

```python
# from pyspark.sql.functions import from_json
# from pyspark.sql.types import StructType,StuctField,StringType
from pyspark.sql.functions import from_json
from pyspark.sql.types import StructType, StructField, StringType

StructTypeschema = StructType([
    StructField("firstname", StringType(),True),
    StructField("lastname", StringType(),True)])
df1 = df.withColumn("nameStruct", from_json(df.name, StructTypeschema))
df1.show(truncate=False)
df1.printSchema()
```

```python
df2=df1.withColumn("firstname",df1.nameStruct.firstname) \
    .withColumn("lastname",df1.nameStruct.lastname)
df2.show(truncate=False)
df2.printSchema()
```
