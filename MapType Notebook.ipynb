# MapType Function

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("MapType").getOrCreate()
```

```python
from pyspark.sql.types import StructType,StructField,StringType,MapType
data = [("balaji",{"village":"darati","dist":"yavatmal"}),("ramkishan",{"village":"chikhali","dist":"nanded"})]

schema = StructType([
    StructField('name', StringType(), True),
    StructField('address', MapType(StringType(),StringType()), True)
])
df=spark.createDataFrame(data,schema)
display(df)
df.printSchema()
```

### access the maptype data

```python
df=df.withColumn("dist",df.address["dist"])
df.show(truncate=False)
```

```python
df=df.withColumn("dist",df.address.getItem("dist"))
df.show(truncate=False)
```

### explode function

```python
from pyspark.sql.functions import explode
df=df.select("name","address",explode(df.address))
df.show(truncate=False)
```

### map_keys

```python
from pyspark.sql.functions import map_keys
df=df.withColumn("keys",map_keys(df.address))
df.show(truncate=False)
```

### map_values

```python
from pyspark.sql.functions import map_values
df=df.withColumn("keys",map_values(df.address))
df.show(truncate=False)
```
